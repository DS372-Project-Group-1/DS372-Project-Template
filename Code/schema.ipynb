```python
import pandas as pd
import numpy as np
from graphviz import Digraph
```

```python
wild = pd.read_csv('https://github.com/MajdBa7r/AQI/raw/main/Fire_Disturbance_Point%20(1).csv')
weather = pd.read_csv("https://github.com/LamaEmran/Data/raw/main/All_Cities_Weather_Final.csv")
aqi = pd.read_csv('https://github.com/MajdBa7r/AQI/raw/main/PM2.5.csv')


def schema_text(df):
    return "\n".join([f"{col} ({str(df[col].dtype)})" for col in df.columns])

wild_schema = schema_text(wild)
weather_schema = schema_text(weather)
aqi_schema = schema_text(aqi)


dot = Digraph(comment="ERD", format="png")
dot.attr(rankdir='LR', fontsize='10')

dot.node("wild", label=f"WILDFIRE\n-------------------\n{wild_schema}", shape="box")
dot.node("weather", label=f"WEATHER\n-------------------\n{weather_schema}", shape="box")
dot.node("aqi", label=f"AQI\n-------------------\n{aqi_schema}", shape="box")


dot.edge("wild", "weather", label="city + date")
dot.edge("weather", "aqi", label="city + date")

dot.render("ERD_diagram2", view=False)
```
